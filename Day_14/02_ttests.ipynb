## One-Sample t-test

```python
# one-sample t-test

#import libraries
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_1samp
```

```python
# load dataset
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df1 = df[['sex', 'age', 'fare']]
```

```python
df1.head()
```

```python
# data key description
df1.describe()
```

```python
# check the age and compare with a known value with 45 years

ttest_1samp(df1['fare'], 50)
stat , p = ttest_1samp(df1['fare'], 50)

print('stat=%.3f, p=%.3f' %(stat,p))

#adding conditional argument for ease
if p>0.05:
    print("probablt the same distribution")
else:
    print('probably diffrent distribution')
```

## 2.Two sample t-test 
# Independent student's t-test

```python
# we will compare age and fare of male vs female passengers


# splitting datasets
df_male = df1.loc[df1['sex']=='male']
df_female = df1.loc[df1['sex']=='female']

# library
from scipy.stats import ttest_ind
stat , p = ttest_ind(df_male['age'], df_female['age'])

print('stat=%.3f, p=%.3f' % (stat,p))

#adding conditional argument for ease
if p > 0.05:
    print("Probably the same distribution")
else:
    print('Probably different distribution')
```

```python
df_male.describe()
```

```python
df_female.describe()
```

```python
# splitting datasets and dropping missing values
df_male = df1.loc[df1['sex']=='male'].dropna(subset=['age'])
df_female = df1.loc[df1['sex']=='female'].dropna(subset=['age'])

# conducting t-test
from scipy.stats import ttest_ind
stat, p = ttest_ind(df_male['age'], df_female['age'])

# printing results
print('stat=%.3f, p=%.3f' % (stat,p))
if p > 0.05:
    print("Probably the same distribution")
else:
    print('Probably different distribution')
```
